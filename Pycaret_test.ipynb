## df merge

```python
## login file too large, need to split
login_df_1 = login_df.loc[(login_df['userid'] >= 0) & (login_df['userid'] <= 100000)]
login_df_2 = login_df.loc[(login_df['userid'] >= 100001) & (login_df['userid'] <= 200000)]
login_df_3 = login_df.loc[(login_df['userid'] >= 200001) & (login_df['userid'] <= 300000)]
login_df_4 = login_df.loc[(login_df['userid'] >= 300001) & (login_df['userid'] <= 400000)]
login_df_5 = login_df.loc[login_df['userid'] >= 400001]
```

```python
login_df_1['date_list'] = login_df_1['date']
login_df_2['date_list'] = login_df_2['date']
login_df_3['date_list'] = login_df_3['date']
login_df_4['date_list'] = login_df_4['date']
login_df_5['date_list'] = login_df_5['date']
```

```python


login_df_1_gb = login_df_1.groupby('userid').agg({'date_list':lambda x:list(x), 'login_times':lambda x:list(x), 'date':max})
login_df_2_gb = login_df_2.groupby('userid').agg({'date_list':lambda x:list(x), 'login_times':lambda x:list(x), 'date':max})
login_df_3_gb = login_df_3.groupby('userid').agg({'date_list':lambda x:list(x), 'login_times':lambda x:list(x), 'date':max})
login_df_4_gb = login_df_4.groupby('userid').agg({'date_list':lambda x:list(x), 'login_times':lambda x:list(x), 'date':max})
login_df_5_gb = login_df_5.groupby('userid').agg({'date_list':lambda x:list(x), 'login_times':lambda x:list(x), 'date':max})
```

```python
login_df_1_gb['s1']=0 ##season1
login_df_1_gb['s2']=0 ##season2
login_df_2_gb['s1']=0 ##season1
login_df_2_gb['s2']=0 ##season2
login_df_3_gb['s1']=0 ##season1
login_df_3_gb['s2']=0 ##season2
login_df_4_gb['s1']=0 ##season1
login_df_4_gb['s2']=0 ##season2
login_df_5_gb['s1']=0 ##season1
login_df_5_gb['s2']=0 ##season2
```

```python
def check_login_count(df):
    for i in range(len(df)):
        date = df.iloc[i].date_list
        login_t = df.iloc[i].login_times
        s1 = 0
        s2 = 0
        for ii,jj in zip(date, login_t):
            if ii < "2020-03-15":
                s1 += jj 
            else:
                s2 += jj
        df.iloc[i, df.columns.get_loc('s1')] = s1
        df.iloc[i, df.columns.get_loc('s2')] = s2
    
        
```

```python
check_login_count(login_df_1_gb)
check_login_count(login_df_2_gb)
check_login_count(login_df_3_gb)
check_login_count(login_df_4_gb)
check_login_count(login_df_5_gb)



```

```python
login_df_final = pd.concat([login_df_1_gb, login_df_2_gb, login_df_3_gb, login_df_4_gb, login_df_5_gb])
```

```python
login_df_final[['s1','s2','date']].to_csv('login_df_final.csv')
```

```python
from pycaret.classification import *
```

```python
train_schedule = setup(data = data, target = 'label', categorical_features = ['gender','is_seller'], numeric_features = [
    's1', 's2', 'shopee_years', 'age', 'order_count', 'total_amount', 'freq', 'last_login', 'userid'],
                       session_id=456,
                      normalize = True,
                      transformation = True,
                      ignore_low_variance = True,
                       remove_multicollinearity = True, multicollinearity_threshold = 0.95,
                      bin_numeric_features = ['userid','shopee_years','age'])
```

```python
best_model_top3 = compare_models(n_select = 3)
```

```python
tuned_lightgbm = create_model('lightgbm', fold = 5)
```

```python
tuned_lightgbm = tune_model(tuned_lightgbm, optimize='AUC')
```

```python
catboost = create_model('catboost')
```

```python
tuned_catboost = tune_model(catboost, optimize='AUC')
```

```python

params = {"iterations": np.arange(100, 250, step=50),
          "learning_rate":  [0.03, 0.03, 0.03, 0.03],
          "depth": [10, 15, 20, 25]
          }
tuned_catboost = tune_model(catboost, custom_grid=params, n_iter=20, fold=4, optimize='AUC')
```
